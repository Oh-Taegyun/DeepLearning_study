# part 7. 합성곱 신경망 (CNN, Convolution Neuarl NetWork)
---

합성곱 신경망(CNN)은 음성 인식, 이미지/영상 인식에서 주로 사용되는 신경망이다.  
다차원 배열 데이터를 처리하도록 구성되어 컬러 이미지 같은 다차원 배열 처리에 특화되어 있다.

### 1. 합성곱 신경망의 구조
---
- 합성곱 신경망은 다섯개의 계층으로 이루어진다.  
    - 입력층  
    - 합성곱층  
    - 풀링층  
    - 완전연결층  
    - 출력층  


1. 입력층
    입력층은 문자 그대로 입력 이미지 데이터가 최초로 거치게 되는 계층이다.

    이미지는 < 가로, 세로, 채널(색상) > 으로 이루어진 3차원 데이터이다.  
    이런 데이터를 1차원으로 받아버리면 3차원 속에서의 본질적인 패턴(밀접하게 관련된 픽셀의 경우 어떤 색을 가지더라 등등...)을 죄다 무시하게 뭉게진다.  
    반면 합성곱 계층은 형상을 유지하게 된다.  
    이미지도 3차원 데이터로 입력받으며, 마찬가지로 다음 계층에서도 3차원 데이터로 전달한다.

### 3. conv2d의 진지한 고찰
---

이 함수가 왜 입력을 이렇게 하는지 출력을 뭐로 하는지 잘 모르겠어서 적는다

자세히 분석해보자 

기본적인 합성곱의 연산은 다음과 같다

![conv의원리](./이미지/image_8.jpg)

1. 파이토치에서의 Conv2d의 형태는 다음과 같다.

    - torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=0, dilation=1, groups=1, bias=True, padding_mode='zeros', device=None, dtype=None)

    - 입력 크기(N, C_in, H, W)와 출력(N, C_out, H_out, W_out)을 갖는 계층의 출력 값은 다음과 같이 정확하게 설명될 수 있다.

    - H_out = ( (H + 2P - FH) / S ) + 1

    - W_out = ( (W + 2P - FW ) / S ) + 1

    + parameters 설명

        - in_channels : 입력 채널 수 

        - out_channels : 출력 채널 수 (컨볼루션에 의해 생성된 채널 수)

        - kernel_size : 커널 크기(필터 크기) 만약에 그냥 정수(ex. a)만 입력했을 경우 (a x a)으로 커널크기를 설정하고, 튜플(ex. (a,b))로 입력했을경우 (a,b)로 커널을 설정

        - padding : 패딩 크기를 입력 커널처럼 정수를 입력했느냐 튜플로 입력했느냐에 따라서 크기 설정

        - 나머지 파라미터 정보 : https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html#torch.nn.Conv2d
        #




    + 엥 합성곱을 했는데 왜 out_channels을 설정할 수 있죠? 계산에 의해서 나오는거 아닌가요? 라고 할 수 있는데 
        
        우리는 왜 계층들을 클래스로 만들고 있는지에 대해 다시 생각해 볼 필요가 있다.

        우리는 매개변수의 관리를 편하게 하기 위해 이렇게 계층을 만들어 왔다. 다시말해 계층이 스스로 매개변수를 갖고 그 매개변수를 스스로 관리하게끔 했다.

        그러면 합성곱 계층에서는 스스로 매개변수를 가져야 한다는 것인데 합성곱의 매개변수는 필터이다.

        그렇다

        바로 출력 채널 수를 정해주는 것은 필터를 정해주는 것이나 다름이 없단 것

        ( 필터 갯수(출력채널수), 채널 수, 커널 크기(세로), 커널 크기(가로) ) 이렇게 필터를 생성할 수 있게되고 이 필터는 합성곱 계층에서 스스로 관리가 가능하게 되었다.
